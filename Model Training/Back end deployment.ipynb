```python
df = log_df['message'].str.split(']', expand=True)
df.rename(columns = {0:'date',1:'id',2:'app',3:'dc centre',4:'server',5:'message'}, inplace = True)
df.head()
```

```python
#Text cleanup
corpus=[]
for i in range(len(df)):
    sent = re.sub('[^a-zA-Z]',' ',df['message'][i])
    sent =  sent.lower()
    sent=" ".join([lemmitizer.lemmatize(word) for word in str(sent).split()])
    sent = " ".join([word for word in str(sent).split() if(word not in stop_words)])
    corpus.append(sent)
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=5000)
X  = cv.fit_transform(corpus).toarray()
```
